```python
cont=['Asia','Europe','North America','South America','Africa','Oceania']
for i in cont:
    print(i,':',len(visa[(visa['continent']==i)&(visa['case_status']=='Denied')]))
```

```python
certified,denied=[],[]
for i in cont:
    count_certified=len(visa[(visa['continent']==i)&(visa['case_status']=='Certified')])
    count_denied=len(visa[(visa['continent']==i)&(visa['case_status']=='Denied')])
    certified.append(count_certified)
    denied.append(count_denied)
    
certified,denied
```

```python
pd.DataFrame(zip(certified,denied),columns=['Certified','Denied'],index=visa['continent'].unique())
```
